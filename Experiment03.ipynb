```python
#Determinant of matrix without Numpy
print("Enter the elements of the 3x3 matrix row-wise:")

matrix = []

for i in range(3):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix.append(row)

a, b, c = matrix[0]
d, e, f = matrix[1]
g, h, i = matrix[2]

determinant = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)

print("\nDeterminant of the matrix:")
print(determinant)

```

```python
#Determinant of matrix using Numpy
import numpy as np

print("Enter the elements of 3x3 matrix row-wise:")

elements = []

for i in range(3):
    row = list(map(int, input(f"Enter row {i+1} (elements separated by space): ").split()))
    elements.append(row)

matrix = np.array(elements)

determinant = np.linalg.det(matrix)

print("\nMatrix:")
print(matrix)

print("\nDeterminant of the given matrix is:")
print(determinant)

```

```python
#Inverse of the matrix without using Numpy
print("Enter the elements of the 3x3 matrix row-wise:")

matrix = []

for i in range(3):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix.append(row)

a, b, c = matrix[0]
d, e, f = matrix[1]
g, h, i = matrix[2]

det = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)

print("\nDeterminant of the matrix:")
print(det)

if det == 0:
    print("Inverse of this matrix does not exist")
else:
    inverse = [
        [(e*i - f*h)/det, (c*h - b*i)/det, (b*f - c*e)/det],
        [(f*g - d*i)/det, (a*i - c*g)/det, (c*d - a*f)/det],
        [(d*h - e*g)/det, (b*g - a*h)/det, (a*e - b*d)/det]
    ]

    print("\nInverse of the matrix is:")
    for row in inverse:
        print(row)

```

```python
print("Solve the system:")
print("a1*x + b1*y = c1")
print("a2*x + b2*y = c2")

a1 = float(input("Enter a1: "))
b1 = float(input("Enter b1: "))
c1 = float(input("Enter c1: "))

a2 = float(input("Enter a2: "))
b2 = float(input("Enter b2: "))
c2 = float(input("Enter c2: "))

# Determinant
D = a1*b2 - a2*b1

if D == 0:
    print("\nThe system has no unique solution (Determinant is zero).")
else:
    Dx = c1*b2 - c2*b1
    Dy = a1*c2 - a2*c1

    x = Dx / D
    y = Dy / D

    print("\nSolution of the system:")
    print("x =", x)
    print("y =", y)
```

```python
#Solve a system of 2 Linear Equations using Numpy
import numpy as np

print("Solve the system:")
print("a1*x + b1*y = c1")
print("a2*x + b2*y = c2")

# Input values
a1 = float(input("Enter a1: "))
b1 = float(input("Enter b1: "))
c1 = float(input("Enter c1: "))

a2 = float(input("Enter a2: "))
b2 = float(input("Enter b2: "))
c2 = float(input("Enter c2: "))

# Create coefficient matrix A
A = np.array([[a1, b1],
              [a2, b2]])

# Create constant matrix B
B = np.array([c1, c2])

# Check determinant
det = np.linalg.det(A)

if det == 0:
    print("\nThe system has no unique solution (Determinant is zero).")
else:
    solution = np.linalg.solve(A,B)
    print("\nSolution of the system:")
    print("x =", solution[0])
    print("y =", solution[1])



```
