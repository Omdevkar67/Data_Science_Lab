```python

# calculating mean, median, mode, variance, SD and plotting histogram

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm

# Ask number of values
n = int(input("Enter number of values: "))

# Take values one by one
print("Enter the values:")
data = []

for i in range(n):
    value = float(input())
    data.append(value)

# Convert list to numpy array
data = np.array(data)

# Mean
mean = np.mean(data)

# Median
median = np.median(data)

# Mode
mode_result = stats.mode(data, keepdims=True)
mode = mode_result.mode[0]

# Variance (Population)
variance = np.var(data)

# Standard Deviation (Population)
std_dev = np.std(data)

# Display Results
print("\nResults:")
print("Mean =", mean)
print("Median =", median)
print("Mode =", mode)
print("Variance =", variance)
print("Standard Deviation =", std_dev)

# --------------------------------
# Plot Histogram using same data
# --------------------------------

plt.hist(data,
         bins=n,      # bins based on number of values
         density=True)

# Generate x values
x = np.linspace(min(data),
                max(data),
                100)

# Generate normal curve using calculated mean & std
y = norm.pdf(x,
             mean,
             std_dev)

# Plot curve
plt.plot(x, y)

# Highlight few points on curve (red)
plt.scatter(x[::10],
            y[::10],
            color='red',
            s=40)

# Labels
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.title("Histogram with Normal Distribution Curve")

plt.show()
```

```python
#Simulate a coin toss experiment and compute probabilities.


import numpy as np

# Take user input
tosses = int(input("Enter number of coin tosses: "))

# Simulate coin toss (H = 1, T = 0)
results = np.random.choice([0, 1], size=tosses)

# Count outcomes
heads = np.sum(results == 1)
tails = np.sum(results == 0)

# Compute probabilities
prob_heads = heads / tosses
prob_tails = tails / tosses

# Display results
print("\nResults:")
print("Total Tosses =", tosses)
print("Heads =", heads)
print("Tails =", tails)
print("Probability of Heads =", prob_heads)
print("Probability of Tails =", prob_tails)
```
